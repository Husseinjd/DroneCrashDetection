```python
def dict_variables(filepath,db,export=False):
    try: #sometimes the dataframe is returing a parse error 
        df = pd.read_csv(filepath,names=range(30),low_memory=False)
    except:
        return -1 #broken file
    
    if len(df) == 1:
        return -1 #broken file
    
    df = df.set_index(0).dropna(axis=1, how='all') 
    try:
        components  = df.loc['FMT'][3].str.strip() #not unique 
    except:
        return -1 #data not complete
        
    sum_unique = 0 # sum unique of "component_subcomponent"
    already_visited_list = [] #keep track of the already visited components (components not a unique list)
    for i,c in enumerate(components):
        export_dict = {} #init on every run
        if c in already_visited_list:
            continue
        else:
            already_visited_list.append(c)
        variable_name_list = []
        c_labels= df.loc['FMT'][components == c].values[0][4:18] #getting all the values
        c_labels = [ob for ob in c_labels if not ob is np.nan] #removing nan values 
        c_labels = np.array([str(c).strip() for c in c_labels]) #cleaning the spaces
        sum_unique += len(c_labels)
        variable_name_list = [c+'_'+sub for sub in c_labels]
        for v in variable_name_list:
            if v in db:
                db[v] += 1 
            else:
                db[v] = 1
        #saving the dict objects 
        if export and c != 'FMT':
            #dataframe with only the variable added
            try:
                df_comp = df.loc[c].dropna(axis=1,how='all')
                #drop any columns which has more than 70% of nan values 
                #df_comp.drop(df_comp.loc[:,(df_comp.isnull().sum()  / len(df_comp) > 0.70)], axis=1 ,inplace=True)
            except:
                print('Component not in index: {}'.format(c))
                continue
            #variables are always less than columns in the log files
            cnter=0
            if len(variable_name_list) < len(df_comp.columns):
                #calculate the difference and add dummy columns
                 for i in range(len(df_comp.columns) -  len(variable_name_list)):
                            variable_name_list.append(c+'_'+str(cnter))
                            cnter+=1
            if len(variable_name_list) > len(df_comp.columns):
#                  for i in range(len(variable_name_list) -  len(df_comp.columns)):
#                              df_comp['c_'+str(cnter)]
#                              cnter+=1
                    continue
                                
            df_comp.columns = variable_name_list.copy()
            for col in df_comp:
                    file_name = filepath[filepath.index('/')+1:-4]
                    export_dict[(col,file_name)] = df_comp[col].copy()
                    #pickle the dict to a file in the directory
                    with open('variables_info/'+col+'_' + file_name+ '.pickle', 'wb') as handle:
                        pickle.dump(export_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
        
    
        del export_dict #free memory
        try:
            del df_comp
        except:
            print('Was not created')
    del df
    del components
```

```python
start = time.time()
db = {} #init dict key:variable value = number of times seen
exdict={}
total_num = len(datalist)
for i,filename in enumerate(datalist):
    print('File:{} --  {} / {}'.format(filename,i+1,total_num-1))
    res = dict_variables('ardu_data/'+ filename,db,export=True)
    if res == -1:
        print('File {} is broken'.format(filename))
end = time.time()
print('Running Time: ',end - start)
```
